# **Pasar el json a DataFrame**

```python
d = json.loads(response_data.content.decode('utf-8'))
d['data'][5000]
```

```python
with open(f"corrida_{d['batch_number']}.json", 'w') as jf: 
    json.dump(response_data.json(), jf, ensure_ascii=False, indent=2)
```
